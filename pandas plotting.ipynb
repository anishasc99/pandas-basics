```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ts = pd.Series(np.random.randn(1000),index=pd.date_range('1/10/2018', periods=1000))
```

```python
plt.figure(figsize=(15,15))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
print(df.head(5))
```

```python
df = df.cumsum()
plt.figure(figsize=(15,15))
plt.figure()
df.plot()
```

#### bar plotting

```python
 plt.figure();

df.iloc[5].plot(kind='bar');
```

#### Density plot

```python
ts.plot.kde()
```

#### Scatter matrix

```python
from pandas.plotting import scatter_matrix,andrews_curves

scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde');
```

#### Multiple bar plotting

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot.bar();
```

#### Stacked bar plot

```python
df2.plot.barh(stacked=True);
```

#### Histogram

```python
plt.figure();
df2.plot.hist(alpha=0.5)
```

#### Stacked histogram

```python
plt.figure();

df2.plot.hist(stacked=True, bins=20)
```

#### Box plot

```python
color = {'boxes': 'DarkGreen', 'whiskers': 'DarkOrange', 'medians': 'DarkBlue', 'caps': 'Gray'}
df2.plot.box(color=color)
```

#### Area plotting

```python
df2.plot.area();
```

```python
df2.plot.area(stacked = False);
```
